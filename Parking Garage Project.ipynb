# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
class ParkingGarage:
    def __init__(self, capacity):
        self.capacity = capacity
        self.open_spots = capacity
        self.taken_spots = 0
        self.tickets = []
        self.parking_spaces = []
        self.current_ticket = None

    def display_options(self):
        print("1. Park the car")
        print("2. Check available spots.")
        print("3. Pay for parking.")
        print("4. Leave the garage.")
        print("5. Exit")

    def takeTicket(self):
        if self.open_spots > 0:
            ticket = {"paid": False}
            self.tickets.append(ticket)
            self.parking_spaces.append(ticket)
            self.current_ticket = ticket
            self.open_spots -= 1
            self.taken_spots += 1
            print("Please take your ticket!")
        else:
            print("I'm sorry, there are no available spots.")

    def payForParking(self):
        if self.current_ticket.get("paid"):
            print("Thank you for your payment! You have 15 minutes to leave the parking garage.")
        else:
            payment = input("Parking today will be $15. Enter 'Yes' to make your payment.")
            if payment.lower() == 'yes':
                self.current_ticket["paid"]= True
                print("Thank you for your payment! You have 15 minutes to leave the parking garage.")
            else:
                print("No payment recieved, please make your payment.")
    
    def leaveGarage(self):
        if self.current_ticket["paid"]:
            print("Thank you, have a great day!")
            self.tickets.remove(self.current_ticket)
            self.parking_spaces.remove(self.current_ticket)
            self.current_ticket = None
            self.taken_spots -= 1
        else:
            print("Please make your payment before leaving the parking garage.")
    
    def check_availability(self):
        print(f"Available spots: {self.open_spots}")

    def interactive_menu(self):
        while True:
            self.display_options()
            choice = input("Please enter your choice (1-5): ")
            if choice == '1':
                self.takeTicket()
            elif choice == '2':
                self.check_availability()
            elif choice == '3':
                self.payForParking()
            elif choice == '4':
                self.leaveGarage()
            elif choice == '5':
                print("GoodBye!")
                break
            else:
                print("Invalid response. Please enter a number from 1-5.")


                
        
            
parking_lot = ParkingGarage(20)
parking_lot.interactive_menu()


```
